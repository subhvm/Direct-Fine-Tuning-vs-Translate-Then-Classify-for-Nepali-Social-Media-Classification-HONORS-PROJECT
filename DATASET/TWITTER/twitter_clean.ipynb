```python
#We use this code to read the cleaned twitter dataset and check the distribution of labels and some sample data.
#The dataset has three classes: -1 for negative sentiment, 0 for neutral sentiment, and 1 for positive sentiment. The distribution of labels is as follows:
#-1: negative sentiment (13593 samples)
#0: neutral sentiment (4860 samples)
#1: positive sentiment (14982 samples)  

```

```python
import pandas as pd

# Load the dataset
tw = pd.read_csv("twitter_dataset_original.csv", on_bad_lines="skip")

# Check columns
print("Columns:", tw.columns.tolist())
print("Shape before cleaning:", tw.shape)
print(tw.head())

# Keep only the columns we need
tw = tw[["Tweet", "Label"]].copy()

# Clean text
tw["Tweet"] = tw["Tweet"].astype(str).str.strip()

# Clean labels
tw["Label"] = tw["Label"].astype(str).str.strip()

# Keep only valid 3-class labels
tw = tw[tw["Label"].isin(["-1", "0", "1"])].copy()

# Convert labels to int
tw["label"] = tw["Label"].astype(int)

# Drop empty tweets
tw = tw[tw["Tweet"] != ""].copy()
tw = tw[tw["Tweet"].notna()].copy()

# Rename text column
tw = tw.rename(columns={"Tweet": "text"})

# Add dataset name
tw["dataset"] = "twitter"

# Reset index and create id
tw = tw.reset_index(drop=True)
tw["id"] = ["tw_" + str(i) for i in range(len(tw))]

# Final column order
tw = tw[["id", "text", "label", "dataset"]]

# Show summary
print("\nShape after cleaning:", tw.shape)
print("\nLabel distribution:")
print(tw["label"].value_counts().sort_index())

print("\nSample rows:")
print(tw.head())

# Save cleaned file
tw.to_csv("twitter_3class_clean.csv", index=False, encoding="utf-8")

print("\nSaved file: twitter_3class_clean.csv")
```
