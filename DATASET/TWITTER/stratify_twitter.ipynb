```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load cleaned Twitter data
df = pd.read_csv("twitter_3class_clean.csv")

# Sanity check
print("Full shape:", df.shape)
print("Full label distribution:")
print(df["label"].value_counts(normalize=True).sort_index())

# First split: train 70%, temp 30%
train_df, temp_df = train_test_split(
    df,
    test_size=0.30,
    stratify=df["label"],
    random_state=42
)

# Second split: val 15%, test 15%
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    stratify=temp_df["label"],
    random_state=42
)

# Reset indices
train_df = train_df.reset_index(drop=True)
val_df = val_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)

# Print summary
print("\nTrain shape:", train_df.shape)
print(train_df["label"].value_counts(normalize=True).sort_index())

print("\nVal shape:", val_df.shape)
print(val_df["label"].value_counts(normalize=True).sort_index())

print("\nTest shape:", test_df.shape)
print(test_df["label"].value_counts(normalize=True).sort_index())

# Save split files
train_df.to_csv("twitter_train.csv", index=False, encoding="utf-8")
val_df.to_csv("twitter_val.csv", index=False, encoding="utf-8")
test_df.to_csv("twitter_test.csv", index=False, encoding="utf-8")

# Save id-only split files too (good for reproducibility)
train_df[["id"]].to_csv("twitter_train_ids.csv", index=False)
val_df[["id"]].to_csv("twitter_val_ids.csv", index=False)
test_df[["id"]].to_csv("twitter_test_ids.csv", index=False)

print("\nSaved:")
print("- twitter_train.csv")
print("- twitter_val.csv")
print("- twitter_test.csv")
print("- twitter_train_ids.csv")
print("- twitter_val_ids.csv")
print("- twitter_test_ids.csv")
```
