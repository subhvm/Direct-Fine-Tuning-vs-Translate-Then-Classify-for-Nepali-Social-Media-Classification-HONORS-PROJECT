```python
from sklearn.model_selection import train_test_split
import pandas as pd

df = pd.read_csv("youtube_2class_clean.csv")

print("Full shape:", df.shape)
print("Full label distribution:")
print(df["label"].value_counts(normalize=True).sort_index())

# 70% train, 30% temp
train_df, temp_df = train_test_split(
    df,
    test_size=0.30,
    stratify=df["label"],
    random_state=42
)

# 15% val, 15% test
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    stratify=temp_df["label"],
    random_state=42
)

train_df = train_df.reset_index(drop=True)
val_df = val_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)

print("\nTrain shape:", train_df.shape)
print(train_df["label"].value_counts(normalize=True).sort_index())

print("\nVal shape:", val_df.shape)
print(val_df["label"].value_counts(normalize=True).sort_index())

print("\nTest shape:", test_df.shape)
print(test_df["label"].value_counts(normalize=True).sort_index())

train_df.to_csv("youtube_train.csv", index=False, encoding="utf-8")
val_df.to_csv("youtube_val.csv", index=False, encoding="utf-8")
test_df.to_csv("youtube_test.csv", index=False, encoding="utf-8")

train_df[["id"]].to_csv("youtube_train_ids.csv", index=False)
val_df[["id"]].to_csv("youtube_val_ids.csv", index=False)
test_df[["id"]].to_csv("youtube_test_ids.csv", index=False)

print("\nSaved:")
print("- youtube_train.csv")
print("- youtube_val.csv")
print("- youtube_test.csv")
print("- youtube_train_ids.csv")
print("- youtube_val_ids.csv")
print("- youtube_test_ids.csv")
```
