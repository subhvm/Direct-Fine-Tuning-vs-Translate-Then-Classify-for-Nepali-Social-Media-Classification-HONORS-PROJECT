```python
import pandas as pd

# Load file
yt = pd.read_csv(
    "YOUTUBE_ORIGINAL.csv",
    header=None,
    names=["label", "aspect", "aspect_term", "text"]
)

print("Raw shape:", yt.shape)
print(yt.head())
print("\nLabel values before cleaning:")
print(yt["label"].value_counts(dropna=False))

# Keep only text + label
yt = yt[["text", "label"]].copy()

# Clean text
yt["text"] = yt["text"].astype(str).str.strip()

# Convert labels safely
yt["label"] = pd.to_numeric(yt["label"], errors="coerce")

# Drop bad rows
yt = yt.dropna(subset=["text", "label"]).copy()
yt = yt[yt["text"] != ""].copy()

# Keep only valid binary labels
yt = yt[yt["label"].isin([0, 1])].copy()
yt["label"] = yt["label"].astype(int)

# Add dataset + id
yt["dataset"] = "youtube"
yt = yt.reset_index(drop=True)
yt["id"] = ["yt_" + str(i) for i in range(len(yt))]

# Final column order
yt = yt[["id", "text", "label", "dataset"]]

print("\nCleaned shape:", yt.shape)
print("\nLabel distribution:")
print(yt["label"].value_counts(normalize=True).sort_index())

print("\nSample rows:")
print(yt.head())

# Save cleaned file
yt.to_csv("youtube_2class_clean.csv", index=False, encoding="utf-8")

print("\nSaved: youtube_2class_clean.csv")
```
